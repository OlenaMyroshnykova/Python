**Conjunción (∧) - "Y"- AND** → Ambas deben ser verdaderas para que toda la oración sea verdadera.


```python
hoy_llueve = True
hace_frio = True
frase = hoy_llueve and hace_frio

print(frase)
```

**Disyunción (∨) - "O"- OR ** → Basta que una sea verdadera para que todo sea verdadero.




```python
voy_al_cine = True
voy_a_comer = False
frase = voy_al_cine or voy_a_comer

print(frase)
```

```python
hay_gasolina = True
bateria_cargada = False

#Usamos AND
puede_arrancar = hay_gasolina and bateria_cargada

print(puede_arrancar)

#Usamos OR
tiene_energia = hay_gasolina or bateria_cargada

print(tiene_energia)

#Usamos NOT
tanque_vacio = not hay_gasolina
print(f" El tanque esta lleno? {tanque_vacio}")

```

Ejercicio para verificar cuanta hambre tienes con if-elif-else

```python
hambre = input("Cuanta hambre tienes? Elije una opción: mucha/mas_o_menos/poca ")
if hambre == "mucha":
  print("Come una pizza grande!")
elif hambre == "mas_o_menos":
  print("Come una manzana.")
elif hambre =="poca":
  print("Toma un vaso de agua")
else:
  print("Vete a correr.")

```

**El Acceso a la Montaña Rusa**

Para subir, hay reglas estrictas:

**Seguridad:** Medir más de 140 cm Y no tener problemas cardíacos (obligatorio).

**Vía Rápida:** Si tienes el "Pase VIP", entras pero siempre que midas más de 120 cm por seguridad básica).

```python
estatura = int(input("¿Cuál es tu estatura? (en cm): "))

tiene_pase_vip = input("¿Tienes pase VIP? (si/no): ").lower() == "si"
problemas_cardiacos = input("¿Tienes problemas cardíacos? (si/no): ").lower() == "si"

# Lógica de acceso
if problemas_cardiacos:
    print("No puedes subir por motivos de salud")

elif tiene_pase_vip and estatura >= 120:
    print("¡Puedes subir por la vía rápida!")
elif estatura >= 140:
    print("¡Puedes subir a la montaña rusa!")

else:
    print("No cumples los requisitos para subir.")
```
